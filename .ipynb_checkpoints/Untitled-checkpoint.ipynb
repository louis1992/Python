Matplotlib 사용법
================
1. 간단한 그래프

jupyter에서 outsection에 그래프를 출력 시키기 위해서는 %matplotlib inline 라는 명령어를 사용해야한다.

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure()            # figure() 객체를 생성.
plt.plot([1,2,3,4,5])   # 1차원 데이터
plt.show()              # 그래프를 최종적으로 그리는 함수.
```

2. numpy를 활용하여 그리기

그래프의 데이터를 변수에 담아 표현할 수 있다.

```python
import numpy as np
x = np.arange(0, 12, 0.01)
print ('x:', x)
plt.figure()
plt.plot(x)
plt.show()
```

```python
y = np.sin(x)
plt.figure(figsize=(6,4))    # 그래프 크기 조절
plt.plot(x,y)
plt.show()
```

그래프 배경에 눈금이나 각 축의 레이블을 정해줄 수 있다.

```python
plt.figure(figsize=(6,4))    
plt.plot(x,y)
plt.grid()             # 그리드 적용
plt.xlabel('time')     # x축 이름
plt.ylabel('Value')    # y축 이름
plt.show()
```

```python
plt.figure(figsize=(6,4))    
plt.plot(x,y)
plt.grid()
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')    # 그래프 타이틀
plt.show()
```

```python
dy = np.diff(y)
print (dy)
dy = np.insert(dy, 0, 0)/0.01
print (dy)
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.plot(x, dy)
plt.grid()
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label='sin')
plt.plot(x, dy, label='cos')
plt.grid()
plt.legend()                    # 각 그래프 범례(legend) 표시.
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')
plt.show()
```

그래프의 형태도 수정이 가능한다.

1) 색상 https://matplotlib.org/api/colors_api.html

2) 굵기

3) 범위

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label='sin')
plt.plot(x, dy, 'r', label='cos')    # cos 그래프를 red로 설정.
plt.grid()
plt.legend()                    
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label='sin')
plt.plot(x, dy, 'r', lw=3, label='cos')    # line Width 를 3으로 설정.
plt.grid()
plt.legend()                    
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label='sin')
plt.plot(x, dy, 'r', lw=3, label='cos')    
plt.grid()
plt.legend()                    
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')
plt.xlim(0, 6)                # x축 범위지정
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, label='sin')
plt.plot(x, dy, 'r', lw=3, label='cos')    
plt.grid()
plt.legend()                    
plt.xlabel('time')
plt.ylabel('Value')
plt.title('Sample Graph')
plt.xlim(0, 6)                # x축 범위지정
plt.ylim(0, 1)                # y축 범위지정
plt.show()
```

```python
t = np.arange(0, 5, 0.5)
plt.figure()
plt.plot(t, t, t, t**2, t, t**3)
plt.show()
```

```python
plt.figure()
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
plt.figure()
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
fig1 = plt.figure(1)
pl1 = plt.plot(t, t, 'r--')

fig1 = plt.figure(2)
pl2 = plt.plot(t, t**2, 'bs')

fig1 = plt.figure(3)
pl3 = plt.plot(t, t**3, 'g^')

plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [1, 4, 5, 8, 9, 5]

plt.figure()
plt.plot(x, y, color='green')
plt.show()
```

```python
plt.figure()
plt.plot(x, y, color='green', linestyle='dashed')
plt.show()
```

```python
plt.figure()
plt.plot(x, y, color='green', linestyle='dashed', marker='o')
plt.show()
```

```python
plt.figure()
plt.plot(x, y, color='green', linestyle='dashed', marker='o', markerfacecolor='blue')
plt.show()
```

```python
plt.figure()
plt.plot(x, y, color='green', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```
